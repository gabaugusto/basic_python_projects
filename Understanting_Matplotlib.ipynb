__Matplotlib: Python plotting__

Matplotlib é uma biblioteca de software para criação de gráficos e visualizações de dados em geral, feita para e da linguagem de programação Python e sua extensão de matemática NumPy.

Before it all, we need to install Matplotlib.

_pip install matplotlib_

To start using Matplotlib we starting import it. 

```python
from matplotlib import pyplot as plt
plt.style.use("ggplot")
```

Now we use some data to feed the lib. 

```python
age_x = [1992, 1995, 1997, 2001, 2005, 2011, 2016, 2020, 2032]
```

```python
# Avengers members by the years
avengers_y = [7, 8, 9, 19, 55,  55, 100, 102, 7]

# X-men members by the years
xmen_y = [5, 7, 11, 14, 36, 98, 150, 381, 729]

# Justice League members by the years
jla_dev_y = [7, 50, 7, 78, 7, 98, 7, 1001, 7]

# Teen Titans members by the years
titans_dev_y = [7, 7, 7, 7, 7, 7,  7, 7, 7]
```

```python
plt.plot(age_x, avengers_y, color="k", linestyle="--", marker="o", label="Avengers")
plt.plot(age_x, xmen_y, color="b", marker="o", linewidth=3, label="X-men")
plt.plot(age_x, jla_dev_y, color="r", marker="o", linewidth=3, label="Justice League")
plt.plot(age_x, titans_dev_y, color="y", marker="o", linewidth=3, label="Teen Titans")

plt.xlabel("Years")
plt.ylabel("Members")
plt.title("Members by the years")
plt.grid(True)
plt.legend()
plt.show()
```
